```python
#Defining it in a new way with one F2(z)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv
mη = 0.547
mρ = 0.7755
mπ = 0.13957
m1 = 1.84
g1 = 1.41
g2 = -3.67
m2 = 2.99
h1 = 2.33
h2 = 5.28
```

```python
#Defining p's and F's
s = np.linspace(1,5)
p1 = np.sqrt ((s**2) + ((mη**2)**2) + ((mπ**2)**2) - (2*s*(mη**2)) - (2*(mη**2)*(mπ**2)) - (2*(mπ**2)*s))/ (2*np.sqrt(s))
p2 = np.sqrt ((s**2) + ((mρ**2)**2) + ((mπ**2)**2) - (2*s*(mρ**2)) - (2*(mρ**2)*(mπ**2)) - (2*(mπ**2)*s))/ (2*np.sqrt(s))
#print(p1) 
#print(p2)
def F2(z):
    return ((z**2)/(9 + (3*z) + (z**2)))

print(F2((5*(p1))**2))
print(F2((5*(p2))**2))
plt.plot(s,F2((5*(p1))**2))
plt.plot(s,F2((5*(p2))**2))
```

```python
def ρ1(s,mηsq,mπsq):
    return ((2*(np.sqrt((s**2) + ((mηsq**2)**2) + ((mπsq**2)**2) - (2*s*(mηsq**2)) - (2*(mηsq**2)*(mπsq**2)) - (2*(mπsq**2)*s))/(2*np.sqrt(s)))/ (8*np.pi*np.sqrt(s))))
#print(ρ1(s,mη,mπ))
plt.plot(s,ρ1(s,mη,mπ), 'r-')
def ρ2(s,mρsq,mπsq):
    return ((2*(np.sqrt((s**2) + ((mρsq**2)**2) + ((mπsq**2)**2) - (2*s*(mρsq**2)) - (2*(mρsq**2)*(mπsq**2)) - (2*(mπsq**2)*s))/(2*np.sqrt(s)))/ (8*np.pi*np.sqrt(s))))
#print(ρ2(s,mρ,mπ))
plt.plot(s,ρ2(s,mρ,mπ),'g-')
```

```python
def ρ(s):
    return np.array([[ρ1(s,mη,mπ)*((F2(5*(p1)))**2),0], [0,ρ2(s,mρ,mπ)*((F2(5*(p2)))**2)]])
    
a = ρ(s)[0,0]*ρ(s)[1,1]
#b = a.tolist()
#print(ρ(s))
plt.plot(s,a,'g-')
```

```python
for s in np.linspace(1,5):
    def K(s):
        return ((np.matrix ([[g1**2, g1*g2], [g2*h1, g2**2]])/ (m1**2 - s)) + (np.matrix([[h1**2, h1*h2], [h2*h1, h2**2]]) / (m2**2 - s)))
    print(K(s))
```

```python
for s in np.linspace(1,5):
    def M(s):
        a = ((ρ(s).dot(K(s))))
        return ((np.identity(2)-(1j*a)).dot(K(s)))
    M11 = F2((5*(p1))**2)*M(s)[0,0]*F2((5*(p1))**2)
    print(M(s))    
    plt.plot(np.sqrt(s),M11.imag, 'g.')
   
        
```
